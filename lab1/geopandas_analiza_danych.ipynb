### Przykłady manipulacji geometrią na kilku wartwach

część wspólna INTERSECTION

```python
polygon = gpd.read_file(powiaty, layer='polygon', bbox=bbox,)
powiaty_gdf = gpd.read_file(powiaty, layer='powiaty', bbox=bbox,)
powiaty_int = gpd.overlay(powiaty_gdf, polygon, how='intersection')
powiaty_int.plot()
```

Różnica DIFFERENCE

```python
powiaty_diff = gpd.overlay(powiaty_gdf, polygon, how='symmetric_difference')
powiaty_diff.plot()
```

Podobieństwo IDENTITY

```python
powiaty_ident = gpd.overlay(powiaty_gdf, polygon, how='identity')
powiaty_ident.plot()
```

# Filtrowanie

```python
import os
import folium
from folium.plugins import Draw, MeasureControl
import geopandas as gpd
import json

gdf.crs = "EPSG:4326"
gdf = gpd.read_file('country.geojson')

gdf = gpd.read_file('country.geojson')
uk=gdf.query("GMI_CNTRY=='AFG' or GMI_CNTRY=='DZA'")


m = folium.Map(
    location=[53.4327,14.5483],
    tiles='OpenStreetMap',
    zoom_start=3,
    control_scale=True,
    crs='EPSG3857'
    )

folium.GeoJson(
    gdf, name='geojson',
    ).add_to(m)


folium.GeoJson(
    uk, name='filtr',
    ).add_to(m)
#wtyczka rysowania
Draw(
    position='topleft',
    draw_options={'polyline': {'allowIntersection': False}},
    edit_options={'poly': {'allowIntersection': False}}
).add_to(m)

# wtyczka pomiarów
MeasureControl(
    position='topright', 
    primary_length_unit='meters', 
    secondary_length_unit='kilometers', 
    primary_area_unit='sqmeters', 
    secondary_area_unit='hectares',
    color_options='red',
).add_to(m)

#wtyczka kontroli wartwy
folium.LayerControl().add_to(m)
#wtyczka minimapy
folium.plugins.MiniMap().add_to(m)
m
```
